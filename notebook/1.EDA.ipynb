### 2.2) Dataset Information

* **Clientum**: Unique IDs of customers formed by a sequence of 9 digits, with a total of **10,127 unique customers** in the dataset.  
* **Attrition_Flag (Target Variable)**: Indicates whether the customer is an **Existing Customer** (current customer) or an **Attrited Customer** (churned customer).  
* **Customer_Age**: Represents the age of customers, ranging from **27 to 73 years**.  
* **Gender**: Coded as **F** for Female and **M** for Male.  
* **Dependent_count**: Represents the number of dependents associated with a customer.  
* **Education_Level**: Indicates the educational qualification of customers with **7 unique values** including **High School, Graduate, Uneducated, College, Post-graduate, Doctorate**, and **Unknown** (1,519 customers).  
* **Marital_Status**: Represents the marital status of customers, with **4 unique values**: **Married, Single, Divorced,** and **Unknown** (749 customers).  
* **Income_Category**: Represents the annual income category of the cardholder, including **Less than $40K, $40K - $60K, $60K - $80K, $80K - $120K, $120K+**, and **Unknown** (1,112 customers).  
* **Card_Category**: Indicates the type of credit card, with **4 unique values**: **Blue, Gold, Silver,** and **Platinum**.  
* **Months_on_book**: Represents the number of months a customer has held an account with the bank.  
* **Total_Relationship_Count**: Represents the number of products held by the customer.  
* **Months_Inactive_12_mon**: The number of months a customer has been inactive in the last **12 months**.  
* **Contacts_Count_12_mon**: The number of times a customer has contacted the bank in the last **12 months**.  
* **Credit_Limit**: Represents the credit limit on the credit card owned by the customer.  
* **Total_Revolving_Bal**: Represents the total revolving balance on the credit card.  
* **Avg_Open_To_Buy**: Represents the average available credit line over the last **12 months**.  
* **Total_Amt_Chng_Q4_Q1**: Represents the change in transaction amount from **Q4 to Q1**.  
* **Total_Trans_Amt**: Represents the total transaction amount in the last **12 months**.  
* **Total_Trans_Ct**: Represents the total transaction count in the last **12 months**.  
* **Total_Ct_Chng_Q4_Q1**: Represents the change in transaction count from **Q4 to Q1**.  
* **Avg_Utilization_Ratio**: Represents the average credit card utilization ratio.

## 4) Exploring Data (Visualization)

```python
def plot_distribution_and_boxplot(data, column_name):
    column_data = data[column_name]
    
    fig, axes = plt.subplots(1, 2, figsize=(14, 6))
    plt.subplots_adjust(wspace=0.5)
    fig.suptitle(f"{column_name}", fontsize=17)
    fig.text(0.5, 0.0016, 'Number of dependents', ha='center', fontsize=14)
    fig.text(0.00001, 0.5, 'Frequency', va='center', rotation='vertical', fontsize=14)

    axes[0].hist(column_data, bins=30, color='#1abc9c', edgecolor='black', alpha=0.7)
    axes[0].axvline(np.mean(column_data), color='#8C4F78', linestyle='--', label='Mean')
    axes[0].axvline(np.median(column_data), color='#000000', linestyle='-', label='Median')
    axes[0].set_title("Distribution", fontsize=15)
    axes[0].tick_params(labelsize=14)
    axes[0].legend()
    axes[0].set_xlabel(None)
    axes[0].set_ylabel(None)

    axes[1].boxplot(column_data, vert=False, patch_artist=True, 
                    boxprops=dict(facecolor='#FFC300', color='black'),
                    medianprops=dict(color='#000000', linewidth=2),
                    meanprops=dict(marker='o', markerfacecolor='#8C4F78', markersize=7))
    axes[1].set_title("Outliers Detection", fontsize=15)
    axes[1].tick_params(labelsize=14)
    axes[1].set_xlabel(None)
    axes[1].set_ylabel(None)

    fig.tight_layout()
    fig.show()
```

```python
def plot_characteristics(df, column_name, sign, value):
    # Filter data based on the condition
    if sign == '>=':
        filtered_data = df[df[column_name] >= value]
    elif sign == '<=':
        filtered_data = df[df[column_name] <= value]
    elif sign == "==":
        filtered_data = df[df[column_name] == value]
    else:
        raise ValueError("Sign must be '>' or '<='")
    
    # Create the figure and axes
    fig, axes = plt.subplots(3, 2, figsize=(14, 6))
    plt.subplots_adjust(hspace=0.6, wspace=0.5)
    fig.suptitle(f"Characteristics of the {column_name}", fontsize=17)
    fig.text(0.5, 0.0016, 'Frequency', ha='center', fontsize=14)
    fig.text(0.00001, 0.5, 'Value', va='center', rotation='vertical', fontsize=14)
    sns.set_context('talk')

    # List of columns for which we are generating countplots
    columns = ['Attrition_Flag', 'Gender', 'Card_Category', 'Income_Category', 'Education_Level', 'Marital_Status']
    colors = ['#8C4F63', '#828C4F', '#8C4F63', '#828C4F', '#8C4F63', '#828C4F']
    
    # Generate countplots for each column in the list
    for i, col in enumerate(columns):
        ax = axes[i // 2, i % 2]
        
        # Filter and sort data for each column
        sorted_data = filtered_data[col].value_counts().index
        sns.countplot(y=filtered_data[col], order=sorted_data, color=colors[i], ax=ax)
        
        ax.tick_params(labelsize=7)
        ax.set(xlabel=None, ylabel=None)
        
        # Set title for each plot
        fontsize = 15
        ax.set_title(col.replace('_', ' ').title(), fontdict={'fontsize': fontsize})

    # Adjust layout for better spacing
    fig.tight_layout()
    fig.show()
```

```python
# Set figure size and style
plt.figure(figsize=(12, 6), dpi=80)
sns.set_style("darkgrid", {"axes.facecolor": ".9"})
sns.set_context('poster')

# Total count for percentage calculations
total = len(df['Attrition_Flag'])

# Formatting function for percentage and count
def my_fmt(x):
    return '{:.2f}%\n({:.0f})'.format(x, total * x / 100)

# Pie chart for churn frequency
plt.pie(df['Attrition_Flag'].value_counts().values,
        labels=df['Attrition_Flag'].value_counts().index,
        pctdistance=0.6,
        shadow=True,
        labeldistance=1.1,
        radius=1.5,
        explode=[0.1, 0], 
        colors=['#4E8C77', '#EAD45A'],
        textprops={'fontsize': 14},
        autopct=my_fmt)

# Title and layout adjustments
plt.title('Frequency of Churn', fontsize=17)
plt.axis('equal')  # Ensure pie chart is circular
plt.show()
```

```python
def plot_pie_count_distribution(df, column_name):
    fig, ax = plt.subplots(1, 2, figsize=(20, 6))
    plt.subplots_adjust(wspace=0.7)
    sns.set_style("darkgrid", {"axes.facecolor": ".9"})
    sns.set_context('talk')

    # Total count for percentage calculation
    total = len(df[column_name])

    def my_fmt(x):
        return '{:.2f}%\n({:.0f})'.format(x, total * x / 100)

    # Pie chart for gender distribution
    ax[0].pie(df[column_name].value_counts().values,
              labels=df[column_name].value_counts().index,
              pctdistance=0.5,
              shadow=True,
              labeldistance=1.1,
              radius=1.5,
              colors=['#4E8C77','#C70039','#1abc9c','#DAF7A6','#FFC300','#FF5733','#C70039','#900C3F'],
              startangle=85,
              textprops={'fontsize': 7},
              autopct=my_fmt)

    ax[0].set_title(f'Distribution of {column_name}', fontsize=17)
    ax[0].axis('equal')

    # Count plot for Gender vs Churn
    sns.countplot(data=df, x=column_name, hue="Attrition_Flag",
                  palette=['#4F788C', '#8C4F78'],
                  order=df[column_name].value_counts().index, ax=ax[1])

    plt.setp(ax[1].get_legend().get_title(), fontsize='10')
    plt.setp(ax[1].get_legend().get_texts(), fontsize='10')
    ax[1].legend(bbox_to_anchor=(1, 1))
    ax[1].set(xlabel=None, ylabel=None)
    ax[1].tick_params(labelsize=7)
    ax[1].set_title(f'{column_name} vs Churn', fontsize=17)

    # Adjust layout
    fig.tight_layout()
    plt.show()
```
